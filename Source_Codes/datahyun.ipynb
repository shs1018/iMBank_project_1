```python
import pandas as pd
import matplotlib.pyplot as plt

# 2021, 2022, 2023 파일 불러오기
data_2021 = merged_full_2021
data_2022 = merged_full_2022
data_2023 = merged_full_2023

# 연도별 MZ세대 '일반한식' 소비 금액 계산 함수
def calculate_yearly_spending(data, year):
    # 필터링: 대구광역시, 경상북도 거주 & MZ세대(20대, 30대) & '일반한식' 업종
    filtered_data = data[
        (data['자택_시도'].isin(['대구광역시', '경상북도'])) &
        (data['연령대'].isin(['20대', '30대'])) &
        (data['가맹점업종명'] == '일반한식')
    ]
    # 승인금액 합계 계산
    return filtered_data['승인금액'].sum()

# 연도별 소비 금액 계산
spending_2021 = calculate_yearly_spending(data_2021, 2021)
spending_2022 = calculate_yearly_spending(data_2022, 2022)
spending_2023 = calculate_yearly_spending(data_2023, 2023)

# 연도별 소비 금액 데이터프레임 생성
yearly_spending = pd.DataFrame({
    '연도': [2021, 2022, 2023],
    '소비금액': [spending_2021, spending_2022, spending_2023]
})

# 시각화
plt.figure(figsize=(10, 6))
bars = plt.bar(yearly_spending['연도'], yearly_spending['소비금액'], color='forestgreen')

# y축 범위 조정 (소수점 없는 단위로 표시하기 위해)
plt.ylim(0, max(yearly_spending['소비금액']) * 1.1)

# 금액을 각 막대 위에 표시
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:,.0f}원', ha='center', va='bottom', fontsize=10)

# 그래프 제목과 라벨
plt.title('대구경북 MZ 고객의 연도별 일반한식 소비 금액 (2021-2023)', fontsize=14)
plt.xlabel('연도', fontsize=12)
plt.ylabel('일반한식 소비 금액 (원)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 2021, 2022, 2023 파일 불러오기
data_2021 = merged_full_2021
data_2022 = merged_full_2022
data_2023 = merged_full_2023


# '승인건수'를 기준으로 MZ 고객을 분류하는 함수
def categorize_mz_by_approval_count(approval_count):
    if approval_count == '5번초과 10번이하':
        return '소비잘안하는MZ'
    elif approval_count == '10번초과 15번이하':
        return '소비하는MZ'
    elif approval_count == '15번초과 20번이하':
        return '소비잘하는MZ'
    elif approval_count == '20번 초과':
        return '매우소비잘하는MZ'
    return None  # '기타'는 None으로 처리하여 포함하지 않음

# 특정 연도의 MZ 고객 승인건수 그룹별 비율 계산 함수
def calculate_mz_approval_ratio(data, year):
    # 필터링: 대구광역시, 경상북도 거주 & MZ세대(20대, 30대) & '일반한식' 업종
    filtered_data = data[
        (data['자택_시도'].isin(['대구광역시', '경상북도'])) &
        (data['연령대'].isin(['20대', '30대'])) &
        (data['가맹점업종명'] == '일반한식')
    ]
    
    # '승인건수'를 그룹화하여 새로운 컬럼 추가
    filtered_data['승인건수_그룹'] = filtered_data['승인건수'].apply(categorize_mz_by_approval_count)
    
    # 각 그룹의 비율 계산
    approval_ratio = filtered_data['승인건수_그룹'].value_counts(normalize=True) * 100  # 비율(%)
    
    # 그룹 순서를 고정
    approval_ratio = approval_ratio.reindex(['소비잘안하는MZ', '소비하는MZ', '소비잘하는MZ', '매우소비잘하는MZ']).fillna(0)
    
    return approval_ratio

# 각 연도별 승인건수 그룹 비율 계산
approval_ratio_2021 = calculate_mz_approval_ratio(data_2021, 2021)
approval_ratio_2022 = calculate_mz_approval_ratio(data_2022, 2022)
approval_ratio_2023 = calculate_mz_approval_ratio(data_2023, 2023)

# 연도별 데이터를 하나의 데이터프레임으로 결합
approval_ratios = pd.DataFrame({
    '2021': approval_ratio_2021,
    '2022': approval_ratio_2022,
    '2023': approval_ratio_2023
}).fillna(0)  # 값이 없는 부분은 0으로 채움

# 시각화
approval_ratios.T.plot(kind='bar', stacked=True, figsize=(12, 8), color=['#FFD700', '#ADFF2F', '#00BFFF', '#FF4500'])
plt.title('대구경북 MZ 고객의 연도별 일반한식 승인건수 그룹별 소비 비율 (2021-2023)')
plt.xlabel('연도')
plt.ylabel('소비 비율 (%)')
plt.legend(title='승인건수 그룹', loc='upper left')
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
from scipy.stats import ttest_rel

# card_data 불러오기
#card_data = pd.read_csv('path_to_your_card_data.csv')  # 파일 경로 설정 필요

# 연령대 기준으로 2030세대와 4050세대 데이터 분리
gen_2030 = merged_full_2021[merged_full_2021['연령대'].isin(['20대', '30대'])]
gen_4050 = merged_full_2021[merged_full_2021['연령대'].isin(['40대', '50대'])]

# 각 세대별 가맹점업종명 기준 승인금액 합산
top_2030 = gen_2030.groupby('가맹점업종명')['승인금액'].sum().nlargest(10)
top_4050 = gen_4050.groupby('가맹점업종명')['승인금액'].sum().nlargest(10)

# 동일 업종 추출을 위해 inner join 방식으로 데이터 정리
top_2030_df = pd.DataFrame(top_2030).reset_index()
top_4050_df = pd.DataFrame(top_4050).reset_index()

merged_top = pd.merge(top_2030_df, top_4050_df, on='가맹점업종명', suffixes=('_2030', '_4050'))

# 쌍체 t-검정 수행
t_stat, p_value = ttest_rel(merged_top['승인금액_2030'], merged_top['승인금액_4050'])

# 결과 출력
print("쌍체 t-검정 결과:")
print("t-통계량:", t_stat)
print("p-값:", p_value)

# p-값을 기준으로 두 집단 간 평균 차이 유의 여부 판단
if p_value < 0.05:
    print("유의미한 차이가 존재합니다.")
else:
    print("유의미한 차이가 존재하지 않습니다.")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency


# 연령대 기준으로 2030세대와 4050세대 데이터 분리
gen_2030 = merged_full_2021[merged_full_2021['연령대'].isin(['20대', '30대'])]
gen_4050 = merged_full_2021[merged_full_2021['연령대'].isin(['40대', '50대'])]

# 각 세대별 가맹점업종명 기준 승인건수 합산 및 상위 10개 업종 추출
top_2030 = gen_2030.groupby('가맹점업종명')['승인건수'].sum().nlargest(10)
top_4050 = gen_4050.groupby('가맹점업종명')['승인건수'].sum().nlargest(10)

# 데이터프레임화 후 업종명 기준 병합 (inner join)
top_2030_df = pd.DataFrame(top_2030).reset_index()
top_4050_df = pd.DataFrame(top_4050).reset_index()
merged_top = pd.merge(top_2030_df, top_4050_df, on='가맹점업종명', how='inner', suffixes=('_2030', '_4050'))

# 카이제곱 검정을 위한 빈도 테이블 생성
contingency_table = merged_top[['승인건수_2030', '승인건수_4050']].values

# 카이제곱 독립성 검정 수행
chi2, p, dof, expected = chi2_contingency(contingency_table)

# 결과 출력
print("카이제곱 통계량:", chi2)
print("p-값:", p)
print("자유도:", dof)
print("기대 빈도:\n", expected)

# p-값에 따른 유의성 판단
if p < 0.05:
    print("2030세대와 4050세대의 업종별 방문 빈도에 유의미한 차이가 있습니다.")
else:
    print("2030세대와 4050세대의 업종별 방문 빈도에 유의미한 차이가 없습니다.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data to mimic the structure with '거래년월', '승인금액', and '연령대'
np.random.seed(0)
data = pd.DataFrame({
    '거래년월': pd.date_range(start='2021-01-01', periods=36, freq='ME').to_list() * 4,  # 3 years of data for 4 age groups
    '연령대': np.repeat(['20대', '30대', '40대', '50대'], 36),
    '승인금액': np.random.randint(1000000, 10000000, size=144)  # Random spending amount
})

# Aggregating total spending per month by age group
monthly_spending = data.groupby(['거래년월', '연령대'])['승인금액'].sum().reset_index()

# Pivot the data to have age groups as columns for easier plotting
monthly_spending_pivot = monthly_spending.pivot(index='거래년월', columns='연령대', values='승인금액')

# Plotting multi-line chart for monthly spending trends by age group
plt.figure(figsize=(12, 6))
for age_group in monthly_spending_pivot.columns:
    plt.plot(monthly_spending_pivot.index, monthly_spending_pivot[age_group], label=age_group, marker='o')

# Adding title and labels
plt.title('월별 소비 변동 추세와 연령대별 변화율')
plt.xlabel('거래년월')
plt.ylabel('총 승인금액')
plt.legend(title='연령대')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

# Sample data to mimic the structure with '가맹점업종명_범주화', '승인건수', and '승인금액'
np.random.seed(0)
data = pd.DataFrame({
    '고객ID': np.random.choice(range(1000, 1020), size=100),
    '가맹점업종명_범주화': np.random.choice(['식생활', '서비스_및_여가', '소매_및_생활용품', '교통_및_물류'], size=100),
    '승인건수': np.random.randint(1, 10, size=100),
    '승인금액': np.random.randint(10000, 500000, size=100),
})

# Aggregating data by category to get the total approval count and amount per category
category_spending = data.groupby('가맹점업종명_범주화').agg({'승인건수': 'sum', '승인금액': 'sum'}).reset_index()

# Generating a scatter matrix to analyze relationships between spending and approval counts for each category
plt.figure(figsize=(10, 8))
scatter_matrix(category_spending[['승인건수', '승인금액']], alpha=0.7, diagonal='hist', figsize=(8, 8))
plt.suptitle('업종별 소비 성향 분석을 위한 산점도 매트릭스')
plt.show()
```

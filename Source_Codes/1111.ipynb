```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 예시 (사용자의 실제 데이터로 대체)
data = pd.read_csv('capital_data_1108_pm.csv') # 실제 데이터 경로 사용

# "서울특별시"와 "경기도" 데이터를 필터링
seoul_data = data[data['가맹점_광역시도'] == "서울특별시"]
gyeonggi_data = data[data['가맹점_광역시도'] == "경기도"]

# '가맹점업종명_범주화' 빈도 계산
seoul_categories = seoul_data['가맹점업종명_범주화'].value_counts()
gyeonggi_categories = gyeonggi_data['가맹점업종명_범주화'].value_counts()

# 두 지역의 데이터프레임으로 정리
comparison_df = pd.DataFrame({
    '서울특별시': seoul_categories,
    '경기도': gyeonggi_categories
}).fillna(0)

# 시각화
comparison_df.plot(kind='bar', figsize=(14, 8))
plt.title('서울특별시 vs 경기도 - 가맹점업종명_범주화 빈도 비교')
plt.xlabel('가맹점업종명_범주화')
plt.ylabel('빈도수')
plt.xticks(rotation=45)
plt.legend(title='지역')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 예시 (사용자의 실제 데이터로 대체)
data = pd.read_csv('capital_data_1108_pm.csv') # 실제 데이터 경로 사용

# '서울특별시' 데이터 필터링
seoul_data = data[data['가맹점_광역시도'] == "서울특별시"]

# '가맹점업종명_범주화' 빈도 계산
seoul_categories = seoul_data['가맹점업종명_범주화'].value_counts()

# 시각화: 파이 차트
plt.figure(figsize=(10, 8))
seoul_categories.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='tab20')
plt.title('서울특별시 - 가맹점업종명_범주화 현황')
plt.ylabel('')  # y축 레이블 제거
plt.tight_layout()
plt.show()
```
